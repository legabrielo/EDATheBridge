```python
app =appearances.groupby(["player_pretty_name", "competition_id", "date", "player_id", "game_id"])[["goals", "assists", "minutes_played", "yellow_cards", "red_cards", "player_club_id"]].sum().reset_index()
app

#A ESTA TABLA SE LE DA UN NUEVO NOMBRE Y SE HACE UN RESET INDEX PARA FILTRAR MEJOR Y QUE APAREZCA "JUGADOR": "COMPETITION_ID==ES1"
```

```python
app_es = app[(app["competition_id"]== "ES1") & (app["date"]>= "2014-01")].reset_index().drop(["index"], axis=1)
app_es
# SE APLICA UNA MASCARA PARA VER SOLAMENTE LOS JUGADORES QUE QUEREMOS
```
