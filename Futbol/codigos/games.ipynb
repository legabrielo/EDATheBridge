```python
gam = games.groupby(["game_id", "season", "club_home_pretty_name",  "club_away_pretty_name",'competition_code', "round", "date"])[["home_club_id", "away_club_id", "home_club_goals", "away_club_goals","club_home_pretty_name","round"]].sum().sort_values(by= "season").reset_index()
gam
# SE HACE EL FILTRO PARA VER LOS PARTIDOS POR TEMPORADA, PERO AHORA SOLO DEBERIAMOS DE VER LOS PARTIDOS DE LA ES1 Y LOS NOMBRES DE LOS DOS EEQUIPOS
```

```python
# FILTRAMOS LOS PARTIDOS PARA SOLO TENER LOS DE LA LIGA ESPAÑOLA Y TENEMOS LOS DATOS DESDE EL 2014 HASTA EL 2022 JUNIO <--- falta esto ultimo
gam_es = gam[(gam["competition_code"]== "ES1") & ((gam['date'] >= '2014-08-23') & (gam['date'] <= '2022-05-22'))].reset_index(0).drop(["index"], axis=1).sort_values(by="date")
gam_es
```
